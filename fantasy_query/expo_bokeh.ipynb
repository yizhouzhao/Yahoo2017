```python
from bokeh.plotting import figure, output_file, show, output_notebook
from bokeh.models import NumeralTickFormatter
from bokeh.charts import Donut
from bokeh.layouts import gridplot, row
from bokeh.models import LinearAxis, Range1d
```

```python
his['year'] = np.arange(2010,2017)
```

```python
p = figure(plot_height = 500,plot_width = 700)
p.line(his['year'],his['Total number of users '],color = "navy",alpha = 0.9,line_width = 2, legend = 
       "Total number of users")
p.line(his['year'],his['Number of comissisoners'],color = "firebrick", line_width = 2, legend = 
      "Number of Commissioners")
p.line(his['year'],his['Total number of users ']-his['Number of comissisoners'],color = "darkgreen", line_width = 2, legend = 
      "Number of Players")
p.yaxis.axis_label = 'Count'
p.xaxis.axis_label = "Year"
p.legend.location = "center_right"
p.title.text = "NBA fantasy Users 2010-2016"
p.title.align = "center"
output_notebook()
show(p)
```

```python
p = figure(plot_height = 500,plot_width = 700,y_range=(0,0.8))
p.line(his['year'],his['User returning rate'],color = "navy",alpha = 0.9,line_width = 2, legend = 
       "User retension rate")
#p.line(his['year'],his['Commissioner returning rate'],color = "firebrick", line_width = 2, legend = 
#      "Commissioner to commissioner")
p.line(his['year'],his['Commissioner continuing rate'],color = "grey", line_width = 2, legend = 
      "Commissioner to comissioner")
p.line(his['year'],his['Play to commissioner rate '],color = "orange", line_width = 2, legend = 
      "Player to commissioner")
p.line(his['year'],his['Player returning rate']-his['Play to commissioner rate '],color = "purple", line_width = 2, legend = 
      "Player to player")
p.line(his['year'],his['Commissioner returning rate']-his['Commissioner continuing rate'],color = "black", line_width = 2, legend = 
      "Commissioner to player")
p.yaxis[0].formatter = NumeralTickFormatter(format="0.0%")
p.xaxis.axis_label = "Year"
p.title.text = "NBA fantasy Users Retention 2010-2016"
p.title.align = "center"
output_notebook()
show(p)
```
