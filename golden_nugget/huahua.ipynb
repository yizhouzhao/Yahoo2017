```python
import pandas as pd
import numpy as np
```

```python
indice = {"F":1,"L":2,"A":3,"O":4,"T":5,"N":6}
```

```python
data1 = pd.read_csv("part1.csv")
data2 = pd.read_csv("part2.csv")
data3 = pd.read_csv("part3.csv")
data4 = pd.read_csv("part4.csv")
data5 = pd.read_csv("part5.csv")
```

```python
data = pd.concat([data1,data2,data3,data4,data5])
```

```python
del data1,data2,data3,data4,data5
```

```python
data = data.fillna(0)
data.head()
```

```python
data.columns
```

```python
data.columns = ["rn","begin","end","age","dv1","dv2","bcookie","session","apv","lmail",
                "rmail","signin","follow","tcts","weather",
                     "news","finance","sports","politics","celebrity",
                     "tv","style","more",
                     "promo","search",
                     "scoreboard","homepageset","rapiddwell","nav",
                     "head","stream","streamads",
                'streampolitics', 'streamsports', 'streamus',
       'streamlifestyle', 'streamstyle', 'streamworld',
       'streamhealth', 'streamcelebrity', 'streambusiness',
       'streamentertainment', 'streamtechnology',"comment","love","share"]
```

```python
data['news_day'] = (data['news'] + 0.0)/(data['dv1'])
data['finance_day'] = (data['finance'] + 0.0)/(data['dv1'])
data['sports_day'] = (data['sports'] + 0.0)/(data['dv1'])
data['politics_day'] = (data['politics'] + 0.0)/(data['dv1'])

data['celebrity_day'] = (data['celebrity'] + 0.0)/(data['dv1'])
data['tv_day'] = (data['tv'] + 0.0)/(data['dv1'])
data['style_day'] = (data['style'] + 0.0)/(data['dv1'])
data['more_day'] = (data['more'] + 0.0)/(data['dv1'])
```

```python
data['homepageset_day'] = (data['homepageset'] + 0.0)/(data['dv1'])
```

```python
data['streampolitics_day'] = (data['streampolitics'] + 0.0)/(data['dv1'])
data['streamsports_day'] = (data['streamsports'] + 0.0)/(data['dv1'])
data['streamus_day'] = (data['streamus'] + 0.0)/(data['dv1'])
data['streamhealth_day'] = (data['streamhealth'] + 0.0)/(data['dv1'])
data['streamlifestyle_day'] = (data['streamlifestyle'] + 0.0)/(data['dv1'])
data['streamstyle_day'] = (data['streamstyle'] + 0.0)/(data['dv1'])
data['streamsports_day'] = (data['streamsports'] + 0.0)/(data['dv1'])

data['streamworld_day'] = (data['streamworld'] + 0.0)/(data['dv1'])
data['streamcelebrity_day'] = (data['streamcelebrity'] + 0.0)/(data['dv1'])
data['streambusiness_day'] = (data['streambusiness'] + 0.0)/(data['dv1'])
data['streamentertainment_day'] = (data['streamentertainment'] + 0.0)/(data['dv1'])

data['streamtechnology_day'] = (data['streamtechnology'] + 0.0)/(data['dv1'])
data['comment_day'] = (data['comment'] + 0.0)/(data['dv1'])
data['love_day'] = (data['love'] + 0.0)/(data['dv1'])
data['share_day'] = (data['share'] + 0.0)/(data['dv1'])
```

```python
#mail info
data['mail_day'] = (data['lmail']+data['rmail'] + 0.0)/(data['dv1'])
#
data['tcts_day'] = (data['tcts'] + 0.0)/(data['dv1'])
#
data['weather_day'] = (data['weather'] + 0.0)/(data['dv1'])
#
data['search_day'] = (data['search'] + 0.0)/(data['dv1'])
#
data['promo_day'] = (data['promo'] + 0.0)/(data['dv1'])
#
data['rapiddwell_day'] = (data['rapiddwell'] + 0.0)/(data['dv1'])
#
data['stream_day'] = (data['stream'] + 0.0)/(data['dv1'])
#
data['streamads_day'] = (data['streamads'] + 0.0)/(data['dv1'])
#
data['scrbrd_day'] = (data['scoreboard'] + 0.0)/(data['dv1'])
#
data['strmpolitics_day'] = (data['streampolitics'] + 0.0)/(data['dv1'])
data['cmmt_day'] = (data['comment'] + 0.0)/(data['dv1'])
data['nav_head_day'] = (data['nav'] + data['head']-data['lmail']+0.0)/data['dv1']
```

```python
young = data[data['age'] == 'new']
exist = data[data['age'] == 'existing']
```

```python
from bokeh.io import show
from bokeh.models import (
    ColumnDataSource,
    HoverTool,
    LinearColorMapper,
    BasicTicker,
    PrintfTickFormatter,
    ColorBar,
)
from bokeh.plotting import figure,output_notebook
```

## Mail-Existing Users

```python
size_info = exist.groupby(['begin','end'])['mail_day'].count().unstack()
size_info = size_info[['F',"L","A","O","T","N"]]
size_info = size_info.reindex_axis(sorted(size_info.index, key = lambda x : indice[x]),axis= 0)
mail_info = exist.groupby(['begin','end'])['mail_day'].mean().unstack()
mail_info = mail_info[['F',"L","A","O","T","N"]]
mail_info = mail_info.reindex_axis(sorted(mail_info.index, key = lambda x : indice[x]),axis= 0)


# this is the colormap from the original NYTimes plot
colors = ["#75968f", "#a5bab7", "#c9d9d3", "#e2e2e2", "#dfccce", "#ddb7b1", "#cc7878", "#933b41", "#550b1d"]
mapper = LinearColorMapper(palette=colors, low=mail_info.values.min(), high=mail_info.values.max())

hover1 = HoverTool(tooltips = [
    ("count", "@count"),
    ("click/visiting day","@mailpd")
])

source1 = ColumnDataSource(data = dict(
    x = [1,2,3,4,5,6]*5,
    y = np.repeat([1,2,3,4,5],6),
    count = size_info.values.flatten(),
    mailpd = mail_info.values.flatten()
))
p = figure(tools = [hover1,"save"],x_range = ["F","L","A","O","T","N"],y_range = ["F","L","A","O","T"],
          toolbar_location='left')
opts = dict(x='x', line_color=None, source=source1)
#p.circle(x,y,radius = np.log(size_info.values.flatten())**2/360,fill_alpha = mail_info.values.flatten()
#         ,color = "black",line_color = None)
p.circle(x = 'x',y = 'y',radius = np.log(size_info.values.flatten())**2/360,
        fill_color ={'field':'mailpd','transform':mapper},
        line_color = None,source = source1)



p.title.text = "Mail clicks for existing users"
p.title.align = "center"
p.title.text_font_size = '14pt'
p.yaxis.axis_label = "Week one Jun 22 - Jun 28"
p.xaxis.axis_label = "Week two Jun 29 - Jul 5"
p.xaxis.axis_label_text_font_size = "12pt"
p.yaxis.axis_label_text_font_size = "12pt"

color_bar = ColorBar(color_mapper=mapper, major_label_text_font_size="5pt",
                     ticker=BasicTicker(desired_num_ticks=len(colors)),
                     formatter=PrintfTickFormatter(format="%.2f"),
                     label_standoff=6, border_line_color=None, location=(0, 0))
p.add_layout(color_bar, 'right')
output_notebook()
show(p)
```

```python
#Bokeh Logo Bokeh ABOUT GALLERY DOCS GITHUB


from math import pi
import pandas as pd

from bokeh.io import show
from bokeh.models import (
    ColumnDataSource,
    HoverTool,
    LinearColorMapper,
    BasicTicker,
    PrintfTickFormatter,
    ColorBar,
)
from bokeh.plotting import figure
from bokeh.sampledata.unemployment1948 import data

data['Year'] = data['Year'].astype(str)
data = data.set_index('Year')
data.drop('Annual', axis=1, inplace=True)
data.columns.name = 'Month'

years = list(data.index)
months = list(data.columns)

# reshape to 1D array or rates with a month and year for each row.
df = pd.DataFrame(data.stack(), columns=['rate']).reset_index()

# this is the colormap from the original NYTimes plot
colors = ["#75968f", "#a5bab7", "#c9d9d3", "#e2e2e2", "#dfccce", "#ddb7b1", "#cc7878", "#933b41", "#550b1d"]
mapper = LinearColorMapper(palette=colors, low=df.rate.min(), high=df.rate.max())

source = ColumnDataSource(df)

TOOLS = "hover,save,pan,box_zoom,reset,wheel_zoom"

p = figure(title="US Unemployment ({0} - {1})".format(years[0], years[-1]),
           x_range=years, y_range=list(reversed(months)),
           x_axis_location="above", plot_width=900, plot_height=400,
           tools=TOOLS, toolbar_location='below')

p.grid.grid_line_color = None
p.axis.axis_line_color = None
p.axis.major_tick_line_color = None
p.axis.major_label_text_font_size = "5pt"
p.axis.major_label_standoff = 0
p.xaxis.major_label_orientation = pi / 3

p.rect(x="Year", y="Month", width=1, height=1,
       source=source,
       fill_color={'field': 'rate', 'transform': mapper},
       line_color=None)

color_bar = ColorBar(color_mapper=mapper, major_label_text_font_size="5pt",
                     ticker=BasicTicker(desired_num_ticks=len(colors)),
                     formatter=PrintfTickFormatter(format="%d%%"),
                     label_standoff=6, border_line_color=None, location=(0, 0))
p.add_layout(color_bar, 'right')

p.select_one(HoverTool).tooltips = [
     ('date', '@Month @Year'),
     ('rate', '@rate%'),
]

show(p)      # show the plot
```
